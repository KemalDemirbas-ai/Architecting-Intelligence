### 🧠 The 10-Step Operational Roadmap
Adhering to our disciplined framework to ensure a high-quality "showcase" output:

1.  **Understand the Purpose:** Since we are predicting a continuous numerical value (price), the problem is defined as a **Regression** task.
2.  **Exploratory Data Analysis (EDA):** Utilizing `df.info()` and `df.describe()` to identify anomalies such as missing values or extreme outliers.
3.  **Feature Selection:** Removing redundant ID and index columns like `Unnamed: 0` and `key` to focus on predictive data.
4.  **Type Conversion:** Transforming the `pickup_datetime` string into a usable numerical **'hour'** for temporal analysis.
5.  **Data Wrangling:** Cleaning the dataset by removing negative fares and illogical passenger counts.
6.  **Feature Engineering:** Calculating the **Manhattan Distance** (absolute coordinate differences) to simulate real-world city grid navigation.
7.  **Encoding:** Converting categorical data into numerical format using `pd.get_dummies()` (One-Hot Encoding).
8.  **Data Splitting:** Partitioning the dataset into Training (X_train) and Testing (X_test) sets.
9.  **Model Training:** Initiating the manual "fit-predict" cycle using **Decision Tree, Random Forest, Gradient Boosting,** and **Deep Learning (ANN)**.
10. **Performance Evaluation:** Validating the success using the **R² Score** metric to determine the best financial predictor.

```python
import pandas as pd
import numpy as np
import zipfile
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pickle
import warnings
warnings.filterwarnings("ignore")
```

```python
# --- STEP 2: Read and Inspect Data (EDA) ---
with zipfile.ZipFile('/content/uber (2).zip', 'r') as z:
    z.extractall('/content/uber_project')
df = pd.read_csv('/content/uber_project/uber.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# --- STEP 3: Determine Suitable Columns ---
df = df.drop(['Unnamed: 0', 'key'], axis=1)
```

```python
# --- STEP 4: Convert Categorical to Numeric ---
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
df['hour'] = df['pickup_datetime'].dt.hour
```

```python
# --- STEP 5: Perform Data Manipulation (Clean Data) ---
df.dropna(inplace=True)
df = df[(df['fare_amount'] > 2.5) & (df['fare_amount'] < 100)]
```

```python
# --- STEP 6: Apply Feature Engineering ---
df['distance'] = abs(df['pickup_latitude'] - df['dropoff_latitude']) + abs(df['pickup_longitude'] - df['dropoff_longitude'])
df = df.drop(['pickup_datetime', 'pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude'], axis=1)
```

```python
# --- STEP 7: Apply One-Hot Encoding ---
df = pd.get_dummies(df, drop_first=True)
```

```python
# --- STEP 8: Split Data into X and y ---
x = df.drop('fare_amount', axis=1)
y = df['fare_amount']
```

```python
# --- STEP 9: Train and Predict Models ---
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

results = []
```

```python
#1. Decision Tree Model
dt = DecisionTreeRegressor(random_state=42)
dt.fit(x_train_scaled, y_train)
dt_pred = dt.predict(x_test_scaled)
results.append({"Model": "Decision Tree", "R2_Score": r2_score(y_test, dt_pred)})
```

```python
# 2. Random Forest Model
rf = RandomForestRegressor(n_estimators=50, random_state=42)
rf.fit(x_train_scaled, y_train)
rf_pred = rf.predict(x_test_scaled)
results.append({"Model": "Random Forest", "R2_Score": r2_score(y_test, rf_pred)})
```

```python
# 3. Gradient Boosting Model
gb = GradientBoostingRegressor(random_state=42)
gb.fit(x_train_scaled, y_train)
gb_pred = gb.predict(x_test_scaled)
results.append({"Model": "Gradient Boosting", "R2_Score": r2_score(y_test, gb_pred)})
```

```python
#4. Deep Learning (ANN) Model
dl_model = Sequential()
dl_model.add(Dense(64, activation='relu'))
dl_model.add(Dense(32, activation='relu'))
dl_model.add(Dense(1)) # Single output for regression
dl_model.compile(optimizer='adam', loss='mean_squared_error')

# Training Neural Network
dl_model.fit(x_train_scaled, y_train, epochs=15, batch_size=32, verbose=0)
dl_pred = dl_model.predict(x_test_scaled, verbose=0).flatten()
results.append({"Model": "Deep Learning (ANN)", "R2_Score": r2_score(y_test, dl_pred)})
```

```python
# --- STEP 10: Measure Model Accuracy ---
results_df = pd.DataFrame(results).sort_values(by="R2_Score", ascending=False)
print("\n--- 🚀 FINAL MODEL ACCURACY RESULTS ---\n")
print(results_df.to_string(index=False))
```

```python
# --- STEP 10: Manual Export (Instructor Style) ---
pickle.dump(rf, open('uber_model.pkl', 'wb'))
pickle.dump(scaler, open('uber_scaler.pkl', 'wb'))

print("✅ Files exported as uber_model.pkl and uber_scaler.pkl")
```
